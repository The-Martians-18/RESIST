```python
def plot(img1, img2):
  plt.figure(figsize=(12,12))

  plt.subplot(1,3,1)
  plt.imshow(img1)

  plt.subplot(1,3,2)
  plt.imshow(img2)

  plt.show()
```

```python
img_names = sorted(glob('Martian Inverted Channels' + '/*/images/*.jpg'))
test_img_names = sorted(glob('Test data' + '/*/images/*.jpg'))
images = []
for imgInd in range(len(img_names)):
  img = cv2.imread(img_names[imgInd])
  gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  image_eq = cv2.equalizeHist(gray_image)
  plot(gray_image, image_eq)
  # images.append(np.array(gray_image,dtype=np.uint8))
  images.append(gray_image)
for imgInd in range(len(test_img_names)):
  img = cv2.imread(test_img_names[imgInd])
  gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  image_eq = cv2.equalizeHist(gray_image)
  plot(gray_image, image_eq)
  images.append(gray_image)
```
